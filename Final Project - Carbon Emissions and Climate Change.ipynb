# An Analysis on Causes of Global Climate Change

As shown from the other graphs, climate change has been happening around the globe and increasing over time. While climate change is a normal process of our globe, human development has rapidly progressed it beyond its natural trends. With human development came a major increase in carbon emissions, hastening the speed of climate change. Below is a graphical analysis of carbon emissions over time, from 1980-2020, chronologically and monthly, with a comparison to a graph of rising temperature of the planet.

# Global Carbon Emissions, 1980-2020

This data was obtained from the US Global Change Research Program, at https://data.globalchange.gov/ , showcasing the average global carbon emissions monthly from 1980-2020, a linear but nonetheless steep slope in such a brief time.

The data was obtained and then data framed to both combine the year and month lists from the original dataset and to graph the data with the combined lists.

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd 
from pandas import DataFrame
import matplotlib.dates as mdate

f = open('co2_mm_gl.txt', 'r')
flines = f.readlines()
year = [line.split()[0] for line in flines]
month = [line.split()[1] for line in flines]
average = [line.split()[3] for line in flines]

year_month = {'year': year,
             'month': month,
             'average': average}
df = pd.DataFrame(year_month, columns=['year','month', 'average']) 
#print(df) 

df['Year-Month'] = df['year'].map(str) + df['month'].map(str)

data = {'year-month': df['Year-Month'],
       'average': average}
df2 = pd.DataFrame(data, columns = ['year-month', 'average']) #combines year and month columns, with the average column still there
print(df2)

years_list = ['1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']

plt.figure(figsize = (20, 10))

plt.plot(df['Year-Month'], average)
plt.title('Global Carbon Emissions, 1980-2020')
plt.ylabel('Average Carbon Emission Value')
plt.xlabel('Year')
plt.xticks(np.arange(0, len(year), 12), (years_list))
plt.yticks(np.arange(0, 489, 30))

plt.show()
```

The graph below shows the average global temperature (of land and sea) from the same time frame, yearly, showing a similar slope to the carbon emissions graph above. This similar slope, and the graphs of others in this project show that increasing carbon emissions from humans have directly caused an unnatural global warming.

```python
f1 = open('average_temp.txt', 'r')
f1lines = f1.readlines()
temp_year = [line.split()[0] for line in f1lines]
average_temp = [line.split()[1] for line in f1lines]

plt.figure(figsize = (20,10))
plt.plot(temp_year, average_temp)
plt.title('Global Average Temperature Increase, 1980-2020')
plt.ylabel('Temperature Increase in Celcius (°C)')
plt.xlabel('Year')
```
